```python
import pandas as pd

data = pd.read_csv('2151.csv')
data.head()
```

```python
# 建立 Age 和 pheno_age 的 lm 模型 仿照下面的代码
# model <- lm(pheno_age1 ~ Age.at.recruitment, data = data)
# data$predicted_pheno_age1 <- predict(model)
# data$PhenoAgeAccel <- data$pheno_age1 - data$predicted_pheno_age1
import statsmodels.api as sm

# 添加常数项
data['intercept'] = 1

# 建立线性回归模型
model = sm.OLS(data['pheno_age'], data[['intercept', 'Age']]).fit()

# 预测 pheno_age1
data['predicted_pheno_age'] = model.predict(data[['intercept', 'Age']])

# 计算 PhenoAgeAccel
data['PhenoAgeAccel1'] = data['pheno_age'] - data['predicted_pheno_age']

# 查看结果
data.head()
```

```python
# 我想知道 model 此时的系数和截距，以及模型的统计信息
# 提取回归系数
intercept, age_coef = model.params

# 打印回归系数
print(f"Intercept: {intercept}")
print(f"Age Coefficient: {age_coef}")

# 使用数学公式进行计算
data['predicted_pheno_age'] = intercept + age_coef * data['Age']

# 计算 PhenoAgeAccel
data['PhenoAgeAccel2'] = data['pheno_age'] - data['predicted_pheno_age']

# 查看结果
data.head()
```
